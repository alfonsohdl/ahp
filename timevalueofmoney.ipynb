```python
import matplotlib.pyplot as plt
```

```python
# Cash flows at different time periods
cash_flows = [100, 200, 300, 400]

# Interest rate
interest_rate = 0.05
```

```python
# Calculate Present Value for each time period
present_values = []
accumulated_pv = 0
for t in range(1, len(cash_flows) + 1):
    pv = cash_flows[t - 1] / (1 + interest_rate) ** t
    accumulated_pv += pv
    present_values.append(accumulated_pv)
```

```python
# Plot the Present Value graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cash_flows) + 1), present_values, marker='o')
plt.xlabel('Time Period')
plt.ylabel('Present Value')
plt.title('Present Value of Cash Flows')
plt.grid(True)
plt.show()
```

```python
# Print the numerical values of Cash Flows and Present Values
print("Time Period\tCash Flow\tPresent Value")
print("----------------------------------------")
for t, cash_flow, pv in zip(range(1, len(cash_flows) + 1), cash_flows, present_values):
    print(f"{t}\t\t{cash_flow}\t\t{pv:.2f}")
```

As we can see, the Present Value of the cash flows starts at $100 at t=1 and increases over time. At each time step, we add the discounted value of the cash flow at that point to the accumulated PV. The growth is not linear because the impact of discounting becomes more significant as time goes on.

- Here are the key takeaways from this graph:

1) The Present Value of future cash flows decreases as time goes on due to the effect of discounting.
2) The steeper the curve, the faster the Present Value decreases, highlighting the importance of time and interest rates.
3) The graph illustrates the fundamental principle that money received earlier is worth more than the same amount received later."

```python
# Calculate Future Value and store values for each time period
future_values = []
fv = 0
for t, cash_flow in enumerate(cash_flows, start=1):
    fv += cash_flow * (1 + interest_rate) ** t
    future_values.append(fv)
```

```python
# Print the Future Value
print("Time Period\tCash Flow\tFuture Value")
print("----------------------------------------")
for t, cash_flow, fv in zip(range(1, len(cash_flows) + 1), cash_flows, future_values):
    print(f"{t}\t\t{cash_flow}\t\t{fv:.2f}")
```

```python
# Plotting the Future Value versus time
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cash_flows) + 1), future_values, marker='o')
plt.xlabel('Time Period')
plt.ylabel('Future Value')
plt.title('Future Value of Cash Flows Over Time')
plt.xticks(range(1, len(cash_flows) + 1))
plt.grid(True)
plt.show()
```

- This code calculates the future value for each time period, stores the values in a list, and then prints both the numerical values and plots the Future Value versus time using Matplotlib.

- To determine if it's profitable to invest in a new hypothetical business project, we can compare the Present Value (PV) of future cash flows with the initial investment cost. If the PV of cash flows is greater than the initial investment, the project is potentially profitable.

```python
def calculate_present_value(cash_flows, interest_rate):
    pv = 0
    for t, cash_flow in enumerate(cash_flows, start=1):
        pv += cash_flow / (1 + interest_rate) ** t
    return pv

# Initial investment cost
initial_investment = -1000

# Cash flows at different time periods
cash_flows = [300, 400, 500, 600]

# Interest rate
interest_rate = 0.1

# Calculate Present Value of cash flows
pv_cash_flows = calculate_present_value(cash_flows, interest_rate)

# Calculate Present Value of initial investment
pv_initial_investment = initial_investment / (1 + interest_rate) ** len(cash_flows)

# Calculate Net Present Value (NPV)
net_present_value = pv_cash_flows + pv_initial_investment

# Determine if the project is profitable
if net_present_value > 0:
    decision = "Profitable"
else:
    decision = "Not Profitable"

# Print the results
print(f"Present Value of Cash Flows: {pv_cash_flows:.2f}")
print(f"Present Value of Initial Investment: {pv_initial_investment:.2f}")
print(f"Net Present Value (NPV): {net_present_value:.2f}")
print(f"Decision: {decision}")
```

- In this code, we calculated the Present Value of both the future cash flows and the initial investment. Then, we calculated the Net Present Value (NPV) by summing the two. Finally, we determined if the project is profitable based on the sign of the NPV. If the NPV is positive, the project is considered profitable; if it's negative, the project is not considered profitable.
